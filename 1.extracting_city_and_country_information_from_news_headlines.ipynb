```python
import pandas as pd
import re
import geonamescache
from unidecode import unidecode
import tqdm
```

```python
gc = geonamescache.GeonamesCache()
```

```python
all_countries_name = [unidecode(country['name']) for country in gc.get_countries().values()]

all_cities_name = [unidecode(city['name']) for city in gc.get_cities().values()]
# alternative_names = {unidecode(city['name']):city['name'] for city in gc.get_cities().values() if city['name']!=unidecode(city['name'])}
```

```python
r_cities_name = r"\b"+r"\b|\b".join(sorted(all_cities_name,key=lambda x:len(x),reverse=True))
r_countries_name = r"\b"+r"\b|\b".join(sorted(all_countries_name,key=lambda x:len(x),reverse=True))
```

```python
with open("./data/headlines.txt") as f:
    headlines = [unidecode(text.strip()) for text in f]
df = pd.DataFrame(headlines,columns=['headline'])
df.head()
```

```python
def find_country(x):
    temp = unidecode(x)
    match = re.search(r_countries_name,temp)
    if match:
        return x[match.start():match.end()]
    else:
        return float('NaN')
def find_city(x):
    temp = unidecode(x)
    match = re.search(r_cities_name,temp)
    if match:
        return x[match.start():match.end()]
    else:
        return float('NaN')
```

```python
df["countries"] = df["headline"].apply(lambda x:find_country(x))
df.head(10)
```

```python
df["cities"] = df["headline"].apply(lambda x:find_city(x))
df.head(10)
```
